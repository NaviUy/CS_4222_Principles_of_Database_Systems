```python
# #attributes
# U = {"A", "B", "C", "D", "E"}

# #DEPENDENCIES
# F = {"A": "D", "BC": "AD", "E": "AD", "C": "B"}

U = {"A", "B", "C", "D", "E", "F", "G", "H"}
F = {"A": "BC", "B": "CFG", "E": "A", "F":"EG", "CH": "G"}
```

```python
def findingClosure(key, leftList, rightList):
    #print("Closure for", key, "is:")
    S = set()
    
    if(len(key) > 1):
        S = S.union(set(key))
    else:
        S.add(key)

    for i in range(len(leftList)):
        if leftList[i] in S:
            S.add(rightList[i])
            
    #print("S:", S)
    
    done = False
    previous = set()

    for key in S:
        if(len(key) > 1):
            previous = previous.union(set(list(key)))
        else:
            previous.add(key)

    while(not done):
        for i in range(len(leftList)):
            key = leftList[i]
            if(key in S):
                if(len(rightList[i]) > 1):
                    S = S.union(set(list(rightList[i])))
                else:
                    S.add(rightList[i])
            if(previous == S):
                done = True
            else:
                previous = S
    if(done):
        return S
```

```python
def checkList(set1, set2):
    tempSet = set()
    if(set1 == set2):
        return ("both")
    else:
        if(len(set1) > len(set2)):
            for key in set2:
                if key in set1:
                    tempSet.add(key)
            if(tempSet == set2):
                return ("last")
        elif(len(set2) > len(set1)):
            for key in set1:
                if key in set2:
                    tempSet.add(key)
            if(tempSet == set1):
                return ("first")
        else:
            return ("neither")
            
```

```python
def redundancyCheck(set1, set2):
    #print("WORKING...")
    
    returnSet = list()
    returnSet2 = list()
    
    for i in range(len(set1)):
        tempList = list()
        tempList2 = list()
        
        #print("Inside FOR LOOP")
        for j in range(len(set1)):
            #print("INSIDE ANOTHER FOR LOOP")
            if(j != i):
                #print("Appending:", set1[j], ",", set2[j])
                tempList.append(set1[j])
                tempList2.append(set2[j])
        #print("templist:", tempList)
        #print("templist2:", tempList2)
        #print("closure Key:", set1[i])
        closure = findingClosure(set1[i], tempList, tempList2)
        #print("closure:", closure)
        if set2[i] not in closure:
            #print("not redundant")
            returnSet.append(set1[i])
            returnSet2.append(set2[i])
    
    return ([returnSet, returnSet2])
            
```

```python
def findCandidateKeys(L, M, R, leftList, rightList):
    #IF L gives all of R then L is the candidateKey
    #IF L doesn't try every combination of L + M until it gives you candidate key
    
    tempSet = set()
    currentKeys = list()
    done = False
    
    for i in L:
        tempSet = tempSet.union(findingClosure(i, leftList, rightList))
        currentKeys.append(i)
        
    if tempSet == U:
        done = True
        return currentKeys
    
    tempM = M
    
    #print("initiating M", tempM)
    
    for i in currentKeys:
        tempM.append(i)
        
    #print("adding current keys into temp M", tempM)
        
    lastList = list()
        
    while(not done):
        #check every combination of L and M
        
        #print("current keys in M", tempM)
        
        tempSet = set()
        
        for i in tempM:
            tempSet = tempSet.union(findingClosure(i, leftList, rightList))
            
        #print("tempSet:", tempSet)
        
        if tempSet == U:
            #print("still == U")
            popped = tempM.pop(0)
            #print("popped", popped)
            lastList = tempM
        else:
            #print("last one that works")
            done = True
            return lastList
```
